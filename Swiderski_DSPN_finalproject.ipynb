### Structural fragmentation of brain networks predicts aphasia severity, but not response to treatment. 

### Motivation
Research has shown that the relative (dis)organization of language-related neural networks predicts both aphasia severity and treatment response. For example, Marebwa and colleagues (Marebwa et al., 2017) found that individuals with more severe neural-network disorganization post-stroke had more severe aphasia than individuals with relatively preserved network organization. Furthermore, Bonilha and colleagues (Bonilha, Gleichgerrcht, Nesland, Rorden, & Fridriksson, 2015) found that individual response to phonological/semantic cueing treatments was predicted by measures of temporal-lobe network integration. 

Semantic Feature Analysis (SFA; Boyle & Coelho, 1995), one of the most well-documented impairment-focused aphasia treatments, improves both naming and overall aphasia severity (Efstratiadou, Papathanasiou, Holland, Archonti, & Hilari, 2018; Hula et al., 2017; Oh et al., 2016). However, it remains unknown how residual neural networks of stroke survivors support response to this treatment. The purpose of this study was twofold: To validate that aphasia severity is predicted by global network integration (Marebwa et al., 2017) and assess if temporal-lobe network organization is predictive of response to SFA. 


### Hypotheses:
Following findings from Bonilha et al. (2015) and Marebwa et al (2017), we hypothesized that:
  1. Global network measures including, Normalized Small Worldness (NSW), will predict aphasia severity.
  2. NSW and a local network measure of betweenness centrality from the L-Temporal lobe would predict patient response to treatment. 

### METHODS:
Eighteen participants with aphasia secondary to left-hemisphere stroke > 6 MPO participated. The Comprehensive Aphasia Test (CAT; Swinburn, Porter, & Howard, 2004) was given prior to treatment and aphasia severity was estimated using CAT Modality T-scores. Diffusion spectrum imaging data was completed with a Siemens 3T Tim Trio Scanner using a 2D EPI diffusion sequence and were reconstructed by q-space diffeomorphic reconstruction (Yeh, Wedeen, & Tseng, 2010).  A whole brain connectivity matrix was then generated to estimate Normalized Small Worldness (Bassett & Bullmore, 2017) and average left temporal betweenness centrality (Kintali, 2008) was  estimated as the average BC of BA = 20, 21, 22, 37, 38. 

### TREATMENT:
Participants received intensive SFA treatment (4 hours/day, 4-5days/week, for 4 weeks) as part of a clinical trial at the VA Pittsburgh Healthcare System.

### Variables
  - ##### Betweeness centrality:
      A measure of the extent to which an ROI lies on path between other verticies. Nodes with high BC have a greater influence on the extent information can pass through a network to other nodes. Conversly, removing a node with high BC disrupts the flow of information the greatest in a network. Estimation of BC followed Kintali's algorithm (Kintali, 2008) as follows.  
      
The number of shortest paths between two nodes is equal to:

$$ \lambda_{st} $$

And the number of shortest paths that pass through node v:

$$ \lambda_{st}(v) $$

The fraction of the paths that pass through $ v $ and the total shortest paths

$$ \delta_{st}(v) = \frac{\lambda_{st} (v)}{\lambda_{st}} $$
      

And BC is equal to the sum of the fractoin fo the shortest paths between nodes s and t that pass through node v. 

$$ BC(v)\;=\;\sum_{s,t:s\neq v\neq t} \delta_{st} (v)  $$
      
  - ##### Normalized Small Worldness (NSW; Bassett & Bullmore, 2017): 
A ratio of the total number of shortest paths (L) that include node $ i $ and the number of motifs (triangular node structures  including nodes $ i $ and $ j $
    
    $$ \frac {Path\;length\;L}{Clustering\;Coefficient\;C} $$ 

where the shortest path length (L) is the sum of the shortest paths between nodes i and j multipled by 1 over the total number of possible shortest paths.
    $$ L = \frac{1}{n}\sum l_{ij} $$
    
Clustering coefficient C is equal to the sum of the triangular nodal structures that include $c_{ij}$
  
  $$ C = \frac{1}{N}\sum c_{ij} $$
  
Therefore, NSW $ \sigma $ is equal to:
    
  $$ \sigma = \frac{L}{C} $$
    
  - ##### Months post onset: 
      The time in months from the stroke resulting in aphasia and participation in the study. 
  - ##### Overall aphasia severity: 
    A behavioral measure from the Comprehensive Aphasia Test. This is an estimate that lies within a T distribution.  It is an average T-score from 6 subtests: Spoken language comprehension, written language comprehension, repetition, naming, reading, and writing.
  - ##### Spoken language comprehension: 
    A composite score that lies within a T distribution which is derived from participant performance on  spoken single word, sentence, and paragraph tasks.
  - ##### Naming ability: 
      Number of correct responses on the Philadelphia Naming Test (Roach, Schwartz, Martin, Grewal, & Brecher, 1996). The PNT is a confrontation naming test with 175 items.
  - ##### Response to intervention on :
    Probe data from acquisition (treated) and generalization (untreated) items in the SFA clinical trial.  Data were transformed with empirical logit transforms to reflect the probability of correct responses on the acquisition/generalization probes.

### Analyses
 The primary analyses will be conducted in a Bayesian Framework with linear regression (Stan Development Team, 2017). Further analysis will not be conducted until model fit has been confirmed by the presence of:
  1. Zero divergent transitions. Divergent transitions are when the MCMC chain leaves the parameter space of the observed data and approaches $ \infty $ looking for a new parameter space.
  2. All $\hat R $ are less than or equal to 1.01. Where $\hat R $ is equal to:
  $$ \hat R   = \sqrt{\frac{\hat{var} + \psi|y}{W}} $$
 
where $ \psi|y $ is the models marginal posterior variance and $ W $ is the within sequence variance. As the $n \rightarrow \infty  $, $\hat R $ reduces to one; however, should it be high the statistic suggersts poor model fit beyond repair by increased number of iterations during sampling (Gelman et al., 2013)
  3. Number of Effective samples: Number of effectiv samples (N_Eff in rStan output) quatifies the Markov Chain Monte Carlo estimator accuracey of a given function. When $ \frac{{neff}}{iterations} < 0.001 $  estimates generated from the MCMC algorithm are usually biased. Therefore, the diagnostic test for the number of effective samples will be to insure that $ \frac{{neff}}{iterations} > 0.001. $
  4. Last, generated data from the MCMC function will be plotted against the observed data to insure that the parameter space generated from the models includes that of the observed data.  

### Proportion of Variance explained. The Bayesian $ R^2 $  (Gelman, Goodrich, Gabry, & Vehtari, 2018)
  -Estimation of $ R^2 $ in a Bayesian Framework is slightly different that the traditional eastimation process because estimation with the frequentist formula may produce out of bound estimates.  For example, when model estimates are from too informative or too vague of priors and the estimation of the Total Sum of Squares
  $$ {TSS} = \sum(y-\hat{y} $$ 
  
 is inaccuraetly estimated as less than the Residual Sum of Squares.
 
 $$ {RSS} = \sum(y- \bar{y)} $$
 
 Then:
 
 $$ 0 > R^2 = \frac{{TSS-RSS}}{TSS} $$

Still aimed at estimating the proportion variance explained, the Bayesian $ R^2 $ (Gelman, Goodrich, Gabry, & Vehtari, 2018) is a ratio of the variance of modeled predictive means (${Var}_u $ drawn from a simulated posterior distribution) and the estimated residual variance form the model $({var}_{res}$ Therefore, 

$$ R^2 = \frac{{var}_\mu}{{var}_\mu + {var}_{res}} $$


### R code for The Bayesian $ R^2 $

```r
bayes_R2 <- function(mu, sigma) {
  var_mu <- apply(mu, 1, var)
  sigma2 <- sigma^2
  var_mu / (var_mu + sigma2)
}
```

#### Cross Validation - Pareto Smoopthed Importance Sampling (Gelman et al., 2018, Vehtari, Geman, Goodrich, & Gabry, 2019, Vehtari, Gelman, & Gabry, 2015):

  Next, evaluation of the models performance will be investigated with Leave One Out Cross Validation (LOO). The rational for utilizing LOOCV relative to a test and training set is that this analysis does not include the number of participants necessary for a true out of sample validation technique (James, Witten, Hastie, & Tibshirani, 2017). More specific to Bayesian inference,  a Pareto Smoothed Importance Sampling (PSIS; Vehtari, et al., 2015) approach will be utilized in this analysis. 
  
  This approach truncates skewed weight distributions by fitting a Pareto Distribution to the upper tail to redeuce unstable estimnation  (Vehtari et al., 2015). The parameter of interest for the PSIS LOO will be the estimates of the log liklehood function which is estimted int he bayesian model in the generated quantities block with the name log_lik.

## Estimation of PSISLOO is outlined as follows  (Gelman et al., 2018, Vehtariet al., 2019):

$$ 1- \frac{{Var}_{loo_{res}}}{{Var}_y} $$

where the variance of the modeled residuals is:

$$ {Var}_{loo_{res}} $$

and the variance of model y's predictive means is: 

$$ {{Var}_y} $$

## Bayes Factors
Bayes Factors will be utilized to compare the log liklehood function of the predictor and naive regression models. The bridgesampling package (Gronau, Singmann, & Wagenmakers, 2017) will be utilized to extract the marginal liklehood distribution from each model. The package will then be utilized to compute Bayes Factors. The marginal liklehood is the denominator in bayes theorum and reflects the the probability of observing $y$ given in model $M$.  

### Bayes Theorum:
$$ (\theta|y,M_k) = \frac{(y|\theta,M_k)p(\theta|M_k)}{p(y|M_k)} $$

Note: 
 - y = observed data
 - $\theta$ = model estimated parameters
 - $M_k$ = Model k

Therefore, the Bayes Factor identifies which model has a higher probability of observing the linear predictor, $y$, within the model architectures being compared.  

### Bayes Theorum:
$$ (\theta|y,M_k) = \frac{(y|\theta,M_k)p(\theta|M_k)}{p(y|M_k)} $$

Note: 
 - y = observed data
 - $\theta$ = model estimated parameters
 - $M_k$ = Model k
 
### Estimation of Bayes Factors:
$$ \frac{p(y|M_{naive})}{{p(y|M_k)}}$$

  - Where model $M_k$ is the regressionmodel with the predictor variable.

 - Hamiltonian Sampling conducted with each chain
 - The data block states what the model is expecting to be given.
 - Parameter block provices which parameters will be estimated. Each listed item in the parameter can be given a prior.  Rstan will provide a vague prior if not given.
 - Model: Priors and liklehood function provided.  Standard normal priors given that are not overly informative, but do noty expect estimates greater than 5 SD from the beta mean and 3 from the alpha/sigma means.

  - Generated quanties are estimates that are created based on the parameters estimated in the model.

#### Conclusions:
 - Normalized Small Worldness controlled for about .27% of the variance when it is regressed on Aphasia Severity.
 - The results with respect to aphasia severity are consistent with prior local connectometry analysis of the data set.
     - Fit for this model was acceptable given: 
       1. Convergence Statistics
       2. Posterior predictive checking
       3. PSISLOO estimates under .5 for all estimates.
     - The Bayes factor when comparing a naive model to the NSW linear model predicting aphasia severity was very weak (0.0).  This can be interpreted that the preferred regression model includes NSW as a predictor. 
 - NSW nor BC were predictive of response to treatment.
 - The lack of relationship between the present temporal lobe network measures and patient response to treatment is inconsistent prior local connectometry analysis of the same data set, which demonstrated robust relationship between ventral pathway connectivity and response to this specific treatment. The $R^2$ estimates were overall low ($R^2 < .11$), but even more telling was that the beta coefficients estimated in both the NSW and BC regression models were not credibly different from zero
      - Fit for this model was acceptable given: 
        1. Convergence Statistics
        2. Posterior predictive checking
        3. PSISLOO estimates under .5 for all estimates.
     - Given the data set at hand, BC and NSW are not robust predictors of response to treatment evidenced by$R^2$ estimates less than 0.1.  
     
     
#### Limitations
  - Small sample size (n = 19)
  - Large lesions have major implications on network estimation.
  - Deterministic vs probabilistic fiber tracking: The papers that motivated this study all utilized probalistic and then current study utilized Deterministic.
  - Nonparametric approaches may be more suitable to analzying this data and should be investigated in the future.
      - Eg. Logistic regression for the treatment related outcome variable or LDA in the aphasia severity model.

#### Future directions"
 -  Prior to further data analysis validation of the network measures is needed.  Methods described by Bonilha and colleagues (2015) may not have been fully understood/replicated.  For example, methods were utilized by Bonilha and colleagues (2015) to controlled for the size of each ROI; however, this method was not described in detail therefore difficult to replicate.  
  

### Closing remarks re: creation of a data plan and thoughts on this project.
 Creation of a data plan requires researchers to develop an in depth understanding of the research strategy, goals, independent/dependent variables, and statistical analyses.  Within a Bayesian framework, an oversight I had was how this process needs to be thought about regarding prior selection/tuning.  In retrospect, I should have (and in the future will) dedicate a specific section to priors and utilize a literature search to develop properly informed priors.  In this case, aphasia severity and response to treatment (empirical logit transforms) were the only observed data that I had domain expertise in.  Although not a great weakness to the study, such investigations will aid in a more well throughout analysis and outcomes that may be more reflective of the data's true distributions. 
 In the future, I will also make specific comments and potential alternate data processing pipelines should assumptions or hypotheses not be met. For example, although I do not hypothesize greater predictive utility, the response to treatment models would have been more reflective of the data should I have utilized logistic regression. 
 A strength identified in the creation of this data plan was its ease in presenting errors in data processing.  For example, the initial thought of utilizing only 2000 iterations for the data was not validated in the initial trace plot.  Although easy to remediate, this also presents a very easy way to be transparent about your modifications in data processing to readers.  Although not an error in analysis, this procedure also made it easy to identify errors in the initial posterior predictive checks implemented in the regression model with and respone to treatment .  Here, an error was made by plotting aphasia severity outcomes.   Given that this code was written prior to analysis it easily identified this error, and should the actual data not fit into this posterior predictive check it would have also been easily identified.  This is an approach that I will certainly utilize in the future.
 In closing, although the creation of a data plan from start to finish felt like a lot of work to conduct prior to analyzing the data from my experiment its process proved to be extremely usefil.  It aided in the development of my understanding of the data, the analyses, and the validation procedures.  It also informed me the need to think about alternate approaches in statistical designs should statistical assumptions not be affirmed. Further, it challenged me to think about (albeit in the future) competing hypotheses a priori to statistical analysis should the data not be consistent with your initial hypotheses.
