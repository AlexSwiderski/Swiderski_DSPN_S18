**Answer:**

2. What exactly does it mean to say that a data point is a "random variable is normal with a mean of 100 and standard deviation of 15"?

**Answer:** 
A random variable with mu = 100 and SD = 15 has four main properties.  First, it is assumed to be continuous. Second, it is a symetrical bell shaped distribution. Third, the bell peaks at mu, in this case 100.  Fourth, roughly 65% of the distribution falls within +/- 15 units (one standardized SD) and 95% of the distribution falls within +/- 30 units (2 standarized SDs).

3. Using the *runif* function, run three experiments by simulating the outcomes of rolling a single, six-sided die. Show the distribution of each experiment. Show how the simulated means compare to the expected mean of a fair roll.
        	a) Exp 1: 10 independent throws
        	b) Exp 2: 1,000 independent throws

```r
# -------------------------------
# Question 3
# -------------------------------

library(ggplot2)
library(wesanderson)
#Wesanderson for fun colors based off Wes Anderson movie themes :)

options(repr.plot.width = 4, repr.plot.height =4)

#make some data
Rolls_10 <- as.integer(runif(10, 1, 7))
m10 <- mean(Rolls_10, digits = 5)
print(m10)
Rolls_1k <-as.integer(runif(1000, 1, 7))
m1k <- mean(Rolls_1k, digits = 5)
print(m10k)
Rolls_100k <- as.integer(runif(10000, 1, 7))
m100k <- mean(Rolls_100k, digits = 5)
print(m100k)
fairmean = 3.5

R10 = data.frame(R10 = Rolls_10)
R1k = data.frame(R1k = Rolls_1k)
R100k = data.frame(R100k = Rolls_100k)


#make some plots
#Observed mean of 10 rolls
print(m10)
#Expected fair dice mean
print(fairmean)
ggplot(data = R10) +
  geom_histogram(aes(x = R10), bins = 10, color = "#3B9AB2", fill = "#78B7C5") +  
  geom_segment(aes(x=m10, xend=m10, y = 0, yend=3),
              color="blue", linetype="dashed", size=1) +
  annotate('segment',x=fairmean, xend=fairmean, y = 0, yend=3, alpha =.75, color = "red", size=1) +
  annotate("text", x = .5, y = 3, label = "Observed mean",  hjust = 0, fontface = 2, color = "blue") +
  annotate("text", x = .5, y = 2.875, label = "Fair dice mean",  hjust = 0, fontface = 2, color = "red") +
  ggtitle("One Die 10 throws") + 
  theme(plot.title = element_text(hjust = 0.5),
  axis.title.x = element_text(face = "bold"),
  axis.title.y = element_text(face = "bold"),
  title = element_text(face = "bold"))

#Observed mean of 1000 rolls
print(m10k)
#Expected fair dice mean
print(fairmean)

ggplot(data = R1k) +
  geom_histogram(aes(x = R1k), bins = 10, color = "#EBCC2A", fill = "#E1AF00") +
  geom_segment(aes(x=m1k, xend=m1k, y = 0, yend=175),
               color="blue", linetype="dashed", size=1) +
  annotate('segment', x=fairmean, xend=fairmean, y = 0, yend=175, color = "red", size=1) +
  annotate("text", x = 0.5, y = 180, label = "Observed mean",  hjust = 0, fontface = 2, color = "blue") +
  annotate("text", x = 0.5, y = 175, label = "Fair dice mean",  hjust = 0, fontface = 2, color = "red") +
  ggtitle("One Die 1000 throws") + 
  theme(plot.title = element_text(hjust = 0.5),
  axis.title.x = element_text(face = "bold"),
  axis.title.y = element_text(face = "bold"),
  title = element_text(face = "bold"))

#Observed mean of 10000 rolls
print(m100k)
#Expected fair dice mean
print(fairmean)
ggplot(data = R100k) +
      geom_histogram(aes(x = R100k), bins = 10, color = "#F1BB7B", fill = "#E6A0C4") + 
  geom_segment(aes(x=m100k, xend=m100k, y = 0, yend=1750),
             color="blue", linetype="dashed", size=1) +
  annotate('segment', x=fairmean, xend=fairmean, y = 0, yend=1750, color = "red", size=1) +
  annotate("text", x = 0.5, y = 2000, label = "Observed mean",  hjust = 0, fontface = 2, color = "blue") +
  annotate("text", x = 0.5, y = 1900, label = "Fair dice mean",  hjust = 0, fontface = 2, color = "red") +
  ggtitle("One Die 10000 throws") + 
  theme(plot.title = element_text(hjust = 0.5),
  axis.title.x = element_text(face = "bold"),
  axis.title.y = element_text(face = "bold"),
  title = element_text(face = "bold"))
```

4. Instead of rolling one die, run a set of experiments reporting the outcomes of rolling multiple dice at the same time.
        	a) Exp 1: 10,000 throws, 1 die
        	b) Exp 2: 10,000 throws, 2 dice
        	c) Exp 3: 10,000 throws, 3 dice
        	d) Exp 4; 10,000 throws, 6 dice.

Show the distribution of results for each experiment. Which of these experiments produces a distribution most similar to a normal normal distribution? Justify your conclusion using Q-Q plots.

**Answer:** 

```r
# -------------------------------
# Question 4
# -------------------------------

D1 <- as.integer(runif(10000, 1, 7))
D2 <- as.integer(runif(10000, 1, 7))
D3 <- as.integer(runif(10000, 1, 7))
D4 <- as.integer(runif(10000, 1, 7))
D5 <- as.integer(runif(10000, 1, 7))
D6 <- as.integer(runif(10000, 1, 7))

Dice <- data.frame(One = D1,
                         Two = D1+D2,
                         Three = D1+D2+D3,
                         Six = D1 + D2 + D3 + D4 + D5 +D6)

library(tidyr)
library(reshape2)
DiceRollin <- melt(Dice)

#because normal ggcolors are boring
library(wesanderson)
title <- "Number of Dice"
ggplot(data = DiceRollin, aes(value, fill = variable)) +
  geom_density() +
  scale_fill_manual(title, values = wes_palette("Darjeeling1")) +
  ggtitle("Visulation of 10,000 throws") + 
  theme(plot.title = element_text(hjust = 0.5)) +
  xlab("Value") + ylab("Proportion Observed") + theme(
  axis.title.x = element_text(face = "bold"),
  axis.title.y = element_text(face = "bold"),
  title = element_text(face = "bold"),
  plot.title = element_text(hjust = 0.5))

```

```r
ggplot(data = DiceRollin, aes(sample=value)) + 
         geom_point(alpha = .1, aes(color = variable),stat = "qq") + 
         ggtitle("QQ")  +
  scale_color_manual(title, values = wes_palette("GrandBudapest2")) +
  xlab("Theoretical Normal Distribution") + ylab("Observed Distribution") + theme(
  axis.title.x = element_text(face = "bold"),
  axis.title.y = element_text(face = "bold"),
  title = element_text(face = "bold"),
  plot.title = element_text(hjust = 0.5))
```

As requested, plots of the distributions and their respective Q-Q plots have been generated. Unlike the density plot, the Q-Q plot presents estimates of the observed quantile (from our generated dice rolls) and its respective quantile from a generated distribution of the researchers choosing.  Given that the goal in this question is to assess how normal the distributions are I included the "stat = "qq"" argument within the geom_point command. 

By looking at the distributions it is apparent that rolling one dice 10,000 times is not capapable of producing a normal distribution.  As expected, it has 6 modes which govern the six pink horizontal lines in the figure. However, there are actually 10,000 data points. Thus the long tails are visually presenting that for a large ammount of the data from the theoretical distribution match to single numbers in the observed single dice roll.  This isnt incrediblly suprising given there are only 6 outcomes of the single dice roll and an infinite number of outcomeas between -4 and 4 of the theoretical distribution. 

The two and three dice rolls show a similar multi-modal distribution and similar long tails in the extreme quantiles of the QQ plot.  Though difficult to observe given the large ammount of samples, the horizontal dispersment of each quantile is reducing as the number of dice increases, which eludes to the fact that if enough dice were to be there would be a more linear relationship with a theoreticl normal distribution. This is visually depicted in the coinstrasrt between the dark blue QQ plot reflecting six di rolled 10,000 times relative to the remaining plots.  This can be interpreted that when six dice are thrown there is a greater ammount of options that can be rolled relative to the other experiments, which ultimately allows for the six di roll experiment to be the most normal.
